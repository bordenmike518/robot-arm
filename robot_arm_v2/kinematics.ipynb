```python
def get_rotate_matrix(motor_num, theta, a_len=1):
    if (motor_num == 0):
        R = [[ -np.sin(theta), 0, np.cos(theta), 0],
             [ np.cos(theta), 0, np.sin(theta), 0],
             [0, 1, 0, a_len],
             [0, 0, 0, 1]]
    elif (motor_num == 1):
        R = [[ np.cos(theta), 0, -np.sin(theta), a_len*np.cos(theta)],
             [ np.sin(theta), 0, np.cos(theta), a_len*np.sin(theta)],
             [ 0, -1, 0, 0],
             [0, 0, 0, 1]]
    elif (motor_num == 2):
        R = [[ np.cos(theta), 0, np.sin(theta), 0],
             [ np.sin(theta), 0, -np.cos(theta), 0],
             [0, 1, 0, a_len],
             [0, 0, 0, 1]]
    elif (motor_num == 3):
        R = [[ np.cos(theta), 0, -np.sin(theta), a_len*np.cos(theta)],
             [ np.sin(theta), 0, np.cos(theta), a_len*np.sin(theta)],
             [ 0, -1, 0, 0],
             [0, 0, 0, 1]]
    elif (motor_num == 4):
        R = [[ np.cos(theta), 0, np.sin(theta), 0],
             [ np.sin(theta), 0, -np.cos(theta), 0],
             [ 0, 1, 0, a_len],
             [0, 0, 0, 1]]
    elif (motor_num == 5):
        R = [[ np.cos(theta), -np.sin(theta), 0, a_len*np.cos(theta)],
             [ np.sin(theta), np.cos(theta), 0, a_len*np.sin(theta)],
             [ 0, 0, 1, 0],
             [0, 0, 0, 1]]
    return  R

def degrees_to_radians(degrees):
    return (degrees/180) * np.pi

def dot_product(a, b):
    zip_b = zip(*b)
    # uncomment next line if python 3 : 
    zip_b = list(zip_b)
    return [[sum(ele_a*ele_b for ele_a, ele_b in zip(row_a, col_b)) 
             for col_b in zip_b] for row_a in a]
```
